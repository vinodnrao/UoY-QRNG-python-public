```python
# Python Code for CQDS error rate

import matplotlib.pyplot as plt
import numpy as np

w = np.linspace(0.01,1,100)
r0 = 0.5
r1 = 0.25
r2 = 0.1
r3 = 0.01

e0 = 0.5 * (((w * (1 + r0)) + r0)/(1 + r0 + (2 * w * (1 + r0))))
e1 = 0.5 * (((w * (1 + r1)) + r1)/(1 + r1 + (2 * w * (1 + r1))))
e2 = 0.5 * (((w * (1 + r2)) + r2)/(1 + r2 + (2 * w * (1 + r2))))
e3 = 0.5 * (((w * (1 + r3)) + r3)/(1 + r3 + (2 * w * (1 + r3))))

x0 = -((e0 * np.log2(e0)) + ((1-e0) * np.log2(1-e0)))
x1 = -((e1 * np.log2(e1)) + ((1-e1) * np.log2(1-e1)))
x2 = -((e2 * np.log2(e2)) + ((1-e2) * np.log2(1-e2)))
x3 = -((e3 * np.log2(e3)) + ((1-e3) * np.log2(1-e3)))

I0 = (1- x0 - (w / 2))
I1 = (1- x1 - (w / 2))
I2 = (1- x2 - (w / 2))
I3 = (1- x3 - (w / 2))

plt.figure(dpi=1200)
#plt.plot (e0,I0,label="r = 0.5",linestyle="dotted")
#plt.plot (e1,I1,label="r = 0.25",linestyle="dashdot")
plt.plot (e2,I2,label="r = 0.1",linestyle="solid")
plt.plot (e3,I3,label="r = 0.01",linestyle="dashed")

plt.xlabel(r'error rate', fontsize=12, fontweight='bold')
plt.ylabel(r'I(AB) - I(AE)', fontsize=12, fontweight='bold')
plt.grid()

plt.xlim([0, 0.25])
plt.ylim([0, 0.8])
plt.legend()
#plt.show()

plt.savefig("error_rate.png", dpi = 2000)
plt.savefig("Figure3.pdf", bbox_inches='tight')
```
