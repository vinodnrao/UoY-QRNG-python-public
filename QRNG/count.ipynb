```python
import collections
import pandas as pd

with open(r"output_total.txt", 'r') as fp:
    x = len(fp.readlines())
    print('Total elements:', x)

with open('output_total.txt') as infile:
    counts = collections.Counter(l.strip() for l in infile)
for line, count in counts.most_common():
    print(line, count, count/x)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("output_total.txt", sep=" ", header=None)

x = len(df.axes[0])
print(x)

counts = df.value_counts()
#print(counts)

df2 = pd.DataFrame(counts, columns=list('a'))

prob = counts/x
#print(prob)
df2['b'] = pd.DataFrame(prob)

df2['c'] = -(df2['b'] * np.log2(df2['b']))
df3 = df2.sort_index()
#df3['0'] = df3['0'].apply('{:0>8}'.format)
#print(df2)
print(df3)
total = (np.sum(df2['c'].values))
print(total)

#df2["b"].plot()
df3["b"].plot()
plt.show()
```

```python
import random
from scipy.linalg import toeplitz
import numpy as np
import pandas as pd

x=2.175
y=x*10
z=int(y)

k=120

randomlist = []
for i in range(0,k):
    n = random.randint(0,1)
    randomlist.append(n)
print(randomlist)

tmat_full=toeplitz(randomlist)
#print(tmat_full)
np.savetxt("toeplitz_mat_full.txt",tmat_full, delimiter=",", fmt='%s')

tmat=np.delete(tmat_full, np.s_[z:120], axis=1)
#print(tmat)
np.savetxt("toeplitz_mat.txt",tmat, delimiter=",", fmt='%s')

df_tmat = pd.DataFrame(tmat, index=None)
rows=len(df_tmat.axes[0])
cols=len(df_tmat.axes[1])
print("rows=",rows,", cols=",cols)
```
