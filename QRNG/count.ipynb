```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.linalg import toeplitz

colnames=['a'] 
#df = pd.read_csv("output.txt", names=colnames, header=None, dtype=str)
#df = pd.read_csv("output.txt", sep=" ", header=None)
df = pd.read_csv("rand1.txt", names=colnames, header=None, dtype=str)
print(df)

x = len(df.axes[0])
print(x)

counts = df.value_counts()
#print(counts)

df2 = pd.DataFrame(counts, columns=list('a'))

prob = counts/x
#print(prob)
df2['b'] = pd.DataFrame(prob)

df2['c'] = -(df2['b'] * np.log2(df2['b']))
df3 = df2.sort_index()
r_df3=len(df3.axes[0])
c_df3=len(df3.axes[1])
print("rows=",r_df3,", cols=",c_df3)
#df3['0'] = df3['0'].apply('{:0>8}'.format)
#print(df2)
#print(df3)
total = (np.sum(df2['c'].values))
print(total)

#df2["b"].plot()
#df3["b"].plot()
#plt.show()

#-----------------------------------

#c_tmat=int(total*10)
c_tmat=int(total)
#r_tmat=120
r_tmat=int(input("Enter the size of list : "))

randomlist = []
for i in range(0,r_tmat):
    n = random.randint(0,1)
    randomlist.append(n)
print(randomlist)

tmat_full=toeplitz(randomlist)
#print(tmat_full)
np.savetxt("toeplitz_mat_full.txt",tmat_full, delimiter=",", fmt='%s')

tmat=np.delete(tmat_full, np.s_[c_tmat:r_tmat], axis=1)
print(tmat)
np.savetxt("toeplitz_mat.txt",tmat, delimiter=",", fmt='%s')

df_tmat = pd.DataFrame(tmat, index=None)
r_df_tmat=len(df_tmat.axes[0])
c_df_tmat=len(df_tmat.axes[1])
print("rows=",r_df_tmat,", cols=",c_df_tmat)
#print(df_tmat)

#-----------------------------------
```
